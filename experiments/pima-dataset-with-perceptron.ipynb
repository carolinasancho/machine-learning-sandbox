**Loading Dataset**

```python
import sys
import numpy
sys.path.append('../src/')

import evaluation
import perceptron

pima = numpy.loadtxt('../../Datasets/pima-indians-diabetes.data.txt', delimiter=',')
```

**Dataset Dimensions**

```python
shape(pima)
```

**Ploting two first features of each class**

```python
indices0 = [x for x in xrange(len(pima)) if pima[x][-1] == 0]
indices1 = [x for x in xrange(len(pima)) if pima[x][-1] == 1]
plot(pima[indices0,0], pima[indices0,1], 'go')
plot(pima[indices1,0], pima[indices1,1], 'rx')
show()
```

**Using Perceptron for classification**

```python
# default parameters
p = perceptron.Perceptron()
p.learn(pima[:,:8],pima[:,8:])
predicted = p.predict(pima[:,:8])
#print predicted
confmatrix = evaluation.confusion_matrix(predicted, pima[:,8:], 2)
print confmatrix
```

**Normalization**

```python
pima[:,:8] = pima[:,:8] - pima[:,:8].mean(axis=0)
pima[:,:8] = pima[:,:8] - pima[:,:8].var(axis=0)
p = perceptron.Perceptron()
p.learn(pima[:,:8],pima[:,8:])
predicted = p.predict(pima[:,:8])
#print predicted
confmatrix = evaluation.confusion_matrix(predicted, pima[:,8:], 2)
print confmatrix
```
